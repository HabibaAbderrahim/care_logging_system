## Fill a log file 

```python
def makeFile():
    print('--------')
    ListName =['Hedi' , 'douaa' , 'Houssem']
    ListMat = ['4527TN211','4507TN211','527TN200']
    dd = {}
    for i ,j in zip(ListName,ListMat) :
        dd['carMat']=j
        dd['user']=i

    logging.basicConfig(format='%(asctime)s  %(carMat)s : %(levelname)s %(message)s  %(user)s', 
                    datefmt='%d %b %H:%M:%S' ,
                    level=logging.DEBUG,
                    filename='logging.log')    
    logging.info('IN access' , extra= dd)
 
    logData = {'carMat': '4520TN211' ,'user':'Habiba'}
    logging.info('OUT access' , extra= logData)   

    logData3 = {'carMat': '4527TN211' ,'user':'Habiba'}
    logging.info('OUT access' , extra= logData3)   

    logData5 = {'carMat': '4527TN200' ,'user':'Nour'}
    logging.error('invalid car ID ' , extra= logData5)   

    logData4 = {'carMat': '1507TN211' ,'user':'Habiba'}
    logging.info('OUT access' , extra= logData4)   
 

    logData6 = {'carMat': '0027TN200' ,'user':'Hedi'}
    logging.error('invalid car ID ' , extra= logData6)   

     

    logData8 = {'carMat': '4527TN200' ,'user':'Nour'}
    logging.error('the user Nour is not authorized to access at this time' , extra= logData8)    
    
    logData7 = {'carMat': '4527T0200' ,'user':'Noura'}
    logging.error('invalid car ID ' , extra= logData7) 

    logData2 = {'carMat': '4507TN211' ,'user':'Hamdi'}
    logging.info('IN access' , extra= logData2)   

    logData9 = {'carMat': '000TN200' ,'user':'Yosr'}
    logging.error('the user Yosr is not authorized to access at this time' , extra= logData9)    
    
```
